####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(5, 5))
```

```python
sinasc['APGAR1'].value_counts().sort_index().plot.bar()
plt.title(label="'APGAR1'")
plt.xlabel(xlabel='Apgar primeiro minuto')
plt.tick_params(axis='x', rotation=0)
plt.show()
```

```python
sinasc['APGAR5'].value_counts().sort_index().plot.bar()
plt.title(label="'APGAR5'")
plt.xlabel(xlabel='Apgar quinto minuto')
plt.tick_params(axis='x', rotation=0)
plt.show()
```

```python
sinasc['SEXO'].value_counts().plot.bar()
plt.title(label="'SEXO'")
plt.tick_params(axis='x', rotation=0)
plt.show()
```

```python
sinasc['GESTACAO'].value_counts().reindex(
    ['Menos de 22 semanas', 
     '22 a 27 semanas', 
     '28 a 31 semanas', 
     '32 a 36 semanas', 
     '37 a 41 semanas', 
     '42 semanas e mais']).plot.bar()
plt.title(label="'GESTACAO'")
plt.xlabel(xlabel='Semanas de gestação')
plt.tick_params(axis='x', rotation=45)
plt.show()
```

```python
sinasc['ESTCIVMAE'].value_counts().sort_index().plot.bar()
plt.title(label="'ESTADO_CIVIL_MAE'")
plt.xlabel(xlabel='Estado civil: \n\n1: Solteira \n2: Casada \n3: Viuva \n4: Separado judicialmente/Divorciado \n5: União consensual \n9: Ignorado')
plt.tick_params(axis='x', rotation=0)
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
print('Apgar quinto minuto: \n0 a 10 \n')
print('Consultas de pré-natal: \n1: Nenhuma \n2: de 1 a 3 \n3: de 4 a 6 \n4: 7 e mais \n9: Ignorado')

plt.rc('figure', figsize=(5, 5))
ax = sns.barplot(data=sinasc, 
                 x='APGAR5', 
                 y='CONSULTAS')
ax.set_yticks(ticks=[1, 2, 3, 4])
plt.xlabel(xlabel="'APGAR5': Apgar no quinto minuto")

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(16, 16))
fig, axes = plt.subplots(2, 2)
fig.subplots_adjust(hspace=0.4)

# Gráfico 1
ax1 = sns.countplot(x='SEXO', data=sinasc, ax=axes[0, 0])

# Gráfico 2
ax2 = sns.countplot(x='GESTACAO', 
                    data=sinasc, 
                    order=['Menos de 22 semanas', 
                           '22 a 27 semanas', 
                           '28 a 31 semanas', 
                           '32 a 36 semanas', 
                           '37 a 41 semanas', 
                           '42 semanas e mais'], 
                    ax=axes[0, 1])
ax2.set_xticks(range(len(ax2.get_xticklabels())))  # Definir os ticks explicitamente
ax2.set_xticklabels(labels=ax2.get_xticklabels(), rotation=45)
ax2.set_xlabel(xlabel='Semanas de gestação')

# Gráfico 3
ax3 = sns.countplot(x='ESTCIVMAE', data=sinasc, ax=axes[1, 0])
ax3.set_xticks(range(len(ax3.get_xticklabels())))  # Definir os ticks explicitamente
ax3.set_xticklabels(labels=['1: Solteira', 
                            '2: Casada', 
                            '3: Viúva', 
                            '4: Separado judicialmente\nDivorciado', 
                            '5: União consensual',
                            '9: Ignorado'], 
                    rotation=45)
ax3.set_xlabel(xlabel='Estado civil')

# Gráfico 4
ax4 = sns.countplot(x='CONSULTAS', data=sinasc, ax=axes[1, 1])
ax4.set_xticks(range(len(ax4.get_xticklabels())))  # Definir os ticks explicitamente
ax4.set_xticklabels(labels=['1: Nenhuma', 
                            '2: de 1 a 3', 
                            '3: de 4 a 6', 
                            '4: 7 e mais', 
                            '9: Ignorado'], 
                    rotation=45)
ax4.set_xlabel(xlabel='Número de consultas de pré-natal')

plt.show()

```
