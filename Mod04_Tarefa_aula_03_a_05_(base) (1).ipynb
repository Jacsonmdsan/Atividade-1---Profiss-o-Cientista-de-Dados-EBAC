## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui
df_mat['numero_de_questoes_mat'] = 24
df_mat['percentual_acertos_mat'] = df_mat['Qt_acertos'] / df_mat['numero_de_questoes_mat']

df_por['numero_de_questoes_por'] = 18
df_por['percentual_acertos_por'] = df_por['Qt_acertos'] / df_por['numero_de_questoes_por']

df_geo['numero_de_questoes_geo'] = 8
df_geo['percentual_acertos_geo'] = df_geo['Qt_acertos'] / df_geo['numero_de_questoes_geo']

df_ing['numero_de_questoes_ing'] = 8
df_ing['percentual_acertos_ing'] = df_ing['Qt_acertos'] / df_ing['numero_de_questoes_ing']

df_his['numero_de_questoes_his'] = 8
df_his['percentual_acertos_his'] = df_his['Qt_acertos'] / df_his['numero_de_questoes_his']

df_fis['numero_de_questoes_fis'] = 12
df_fis['percentual_acertos_fis'] = df_fis['Qt_acertos'] / df_fis['numero_de_questoes_fis']

df_qui['numero_de_questoes_qui'] = 12
df_qui['percentual_acertos_qui'] = df_qui['Qt_acertos'] / df_qui['numero_de_questoes_qui']
````

````python
df = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)
df.index += 1
df
````

````python
# 2) Seu código aqui
total_acertos_aluno = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
total_acertos_aluno[['Qt_acertos']] 
````

````python
# 3) Seu código aqui
perc_geral_aluno = total_acertos_aluno[['Qt_acertos']]  / 100
perc_geral_aluno
````

````python
# 4) Seu código aqui
alunos_aprovados = total_acertos_aluno[total_acertos_aluno['Qt_acertos'] > 45]
print(f'A quantidade de alunos aprovados para a segunda fase é {alunos_aprovados['Qt_acertos'].count()}')
````
